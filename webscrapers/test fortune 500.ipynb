```python
def killevil(s, evil):
    bois = s.split(evil[0])
    outs = []
    outs.append(bois[0])
    for boi in bois[1:]:
        splitty = boi.split(evil[1])
        if (len(splitty) > 1):
            outs.append(splitty[1])
        else:
            outs.append(splitty[0])
    return ("".join(outs))

def devourallevil(s, evils):
    guy = s
    for evil in evils:
        guy = killevil(guy, evil)
    return " ".join(guy.split())
```
